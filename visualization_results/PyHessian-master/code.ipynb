```
def plot_top_eigenvalues(path, save_path, layers, models, epochs, model_name, dataset_name, remain_params_file_path):
  # path = '/content/drive/My Drive/PyHessian-master-429/output_fc1_mnist/top_eigenvalues'
  postfixes = ['/']
  for i in range (layers):
    postfixes.append('_layer'+str(i)+'/')

  top_eigenvalues = []
  for i in range(len(postfixes)):
    # print('i = ' + str(i))
    top_eigenvalues.append([])
    files = os.listdir(path + postfixes[i])
    for _ in range(models):
      top_eigenvalues[i].append([0] * epochs)
    # top_eigenvalues = [0] * len(files)
    for file_index in range(len(files)):
      # if file_index % 100 == 0:
      #   print('file_index = ' + str(file_index))
      _model = int(files[file_index][15:18])
      if _model == 0:
        # f.close()
        continue
      _iter = files[file_index][-6:-4]
      if _iter == 'st':
        # f.close()
        continue
      _iter = int(_iter)
      # _index = int(_model * 9 + _iter / 10)
      f = open(path + postfixes[i] + files[file_index], 'r')
      top_eigenvalues[i][_model][int(_iter/10)] = float(f.readline())
      f.close()
  
  remain_params_file = open(remain_params_file_path, 'r')
  remain_params_list = []
  # f = open(path + postfixes[i] + files[file_index], 'r')
  try:
    while True:
      remain_params = remain_params_file.readline()
      if remain_params:
        # print(traces[i])
        # print(_model)
        # print(len(traces[i][0][0]))
        # print(traces[i][_model])
        # print(traces[i][_model][int(_iter/10)])
        remain_params_list.append(float(remain_params))
      else:
        break
  finally:
    remain_params_file.close()
  
  for i in range(len(postfixes)):
    for j in range (1, models):
      # print(j)
      if postfixes[i] == '/':
        folder = 'whole_model'
      else:
        folder = postfixes[i][1:-1]
      plt.plot(range(0, epochs * 10, 10), top_eigenvalues[i][j])
      # remain_params = round(0.8**(j-1), 3) * 100
      remain_params = remain_params_list[j-1]
      plt.title('Top eigenvalues plot for ' + model_name + ' on ' + dataset_name + ' (%.1f%% params remaining)' % remain_params)
      # plt.title('Top EVs plot for ' + model_name + ' on ' + dataset_name + ' (%.1f%%)' % remain_params)
      plt.xlabel('epoch')
      plt.ylabel('value')
      # plt.show()
      if not os.path.exists(save_path + folder):
          os.makedirs(save_path + folder)
      plt.savefig(save_path + folder + '/model' + str(j-1) + '.jpg')
      plt.clf()
```

```
def plot_density_eigen(path, save_path, layers, models, epochs, model_name, dataset_name, remain_params_file_path):
  postfixes = ['/']
  for i in range (layers):
    postfixes.append('_layer'+str(i)+'/')

  density_eigen = []
  for i in range(len(postfixes)):
    # print('i = ' + str(i))
    files = os.listdir(path + postfixes[i])
    density_eigen.append([[[] for col in range(9)] for row in range(31)])
    # top_eigenvalues = [0] * len(files)
    for file_index in range(len(files)):
      # if file_index % 100 == 0:
      #   print('file_index = ' + str(file_index))
      _model = int(files[file_index][13:16])
      if _model == 0:
        # f.close()
        continue
      _iter = files[file_index][-6:-4]
      if _iter == 'st':
        # f.close()
        continue
      _iter = int(_iter)
      # _index = int(_model * 9 + _iter / 10)
      f = open(path + postfixes[i] + files[file_index], 'r')
      try:
        while True:
          text_line = f.readline()
          if text_line:
            # print(traces[i])
            # print(_model)
            # print(len(traces[i][0][0]))
            # print(traces[i][_model])
            # print(traces[i][_model][int(_iter/10)])
            density_eigen[i][_model][int(_iter/10)].append(float(text_line))
          else:
            break
      finally:
        f.close()
    
  remain_params_file = open(remain_params_file_path, 'r')
  remain_params_list = []
  # f = open(path + postfixes[i] + files[file_index], 'r')
  try:
    while True:
      remain_params = remain_params_file.readline()
      if remain_params:
        # print(traces[i])
        # print(_model)
        # print(len(traces[i][0][0]))
        # print(traces[i][_model])
        # print(traces[i][_model][int(_iter/10)])
        remain_params_list.append(float(remain_params))
      else:
        break
  finally:
    remain_params_file.close()

  for i in range(len(postfixes)):
    for j in range (1, models):
      # print(j)
      x = []
      y = []
      if postfixes[i] == '/':
        folder = 'whole_model'
      else:
        folder = postfixes[i][1:-1]
      for k in range(epochs):
        for tmp in density_eigen[i][j][k]:
          x.append(k*10)
          y.append(tmp)
      plt.scatter(x,y)
      # remain_params = round(0.8**(j-1), 3) * 100
      remain_params = remain_params_list[j-1]
      plt.title('SLQ plot for ' + model_name + ' on ' + dataset_name + ' (%.1f%% params remaining)' % remain_params)
      # plt.title('SLQ plot for ' + model_name + ' on ' + dataset_name + ' (%.1f%%)' % remain_params)
      plt.xlabel('epoch')
      plt.ylabel('value')
      # plt.show()
      if not os.path.exists(save_path + folder):
          os.makedirs(save_path + folder)
      plt.savefig(save_path + folder + '/model' + str(j-1) + '.jpg')
      plt.clf()
```

```
def plot_top_eigenvalues_best(path, save_path, layers, models, epochs, model_name, dataset_name, remain_params_file_path):
  # path = '/content/drive/My Drive/PyHessian-master-429/output_fc1_mnist/top_eigenvalues'
  postfixes = ['/']
  for i in range (layers):
    postfixes.append('_layer'+str(i)+'/')

  top_eigenvalues = []
  for i in range(len(postfixes)):
    # print('i = ' + str(i))
    top_eigenvalues.append([])
    files = os.listdir(path + postfixes[i])
    for _ in range(models):
      top_eigenvalues[i].append(0)
    # top_eigenvalues = [0] * len(files)
    for file_index in range(len(files)):
      # if file_index % 100 == 0:
      #   print('file_index = ' + str(file_index))
      _model = int(files[file_index][15:18])
      if _model == 0:
        # f.close()
        continue
      # print(_model)
      _iter = files[file_index][-6:-4]
      if _iter != 'st':
        # f.close()
        continue
      # _iter = int(_iter)
      # _index = int(_model * 9 + _iter / 10)
      f = open(path + postfixes[i] + files[file_index], 'r')
      top_eigenvalues[i][_model] = float(f.readline())
      f.close()
  
  remain_params_file = open(remain_params_file_path, 'r')
  remain_params_list = []
  # f = open(path + postfixes[i] + files[file_index], 'r')
  try:
    while True:
      remain_params = remain_params_file.readline()
      if remain_params:
        # print(traces[i])
        # print(_model)
        # print(len(traces[i][0][0]))
        # print(traces[i][_model])
        # print(traces[i][_model][int(_iter/10)])
        remain_params_list.append(float(remain_params))
      else:
        break
  finally:
    remain_params_file.close()

  for i in range(len(postfixes)):
    # for j in range (1, models):
      # print(j)
    if postfixes[i] == '/':
      folder = 'whole_model'
    else:
      folder = postfixes[i][1:-1]
    plt.plot(range(len(top_eigenvalues[i][1:])), top_eigenvalues[i][1:])
    plt.xticks(range(len(top_eigenvalues[i][1:])), remain_params_list, rotation  = 45)
    # remain_params = round(0.8**(j-1), 3) * 100
    plt.title('Top eigenvalues plot for ' + model_name + ' on ' + dataset_name)
    # plt.xlabel('prune iteration')
    plt.xlabel('remaining params')
    plt.ylabel('value')
    # plt.show()
    
    if not os.path.exists(save_path + folder):
        os.makedirs(save_path + folder)
    plt.savefig(save_path + folder + '/best_model.jpg', bbox_inches = "tight")
    plt.clf()
```
